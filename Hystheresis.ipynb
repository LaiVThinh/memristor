```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("report_table_All.csv")
time = df['time (s)']
current = df['/I0/p (A)']     
voltage = df['/net1 (V)']     

plt.figure(figsize=(8,6))
plt.plot(voltage, current, '-', linewidth=1.5)
plt.xlabel('Voltage (V')
plt.ylabel('Current (A)')
plt.title('Hysteresis loop (V-I)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
